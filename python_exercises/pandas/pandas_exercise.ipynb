```python
import pandas as pd




bookings = pd.read_csv('CSV_DATA/bookings.csv')
facilities = pd.read_csv('CSV_DATA/facilities.csv')
members = pd.read_csv('CSV_DATA/members.csv')

```

```python
#How can you produce a list of the start times for bookings by members named 'David Farrell'?
q_1=bookings.merge(members,on=['memid']).query('firstname == "David"').query('surname == "Farrell"')
q_1['starttime']
```

```python
#How can you produce a list of the start times for bookings for tennis courts, for the date '2012-09-21'? 
#Return a list of start time and facility name pairings, ordered by the time.

q_2=bookings.merge(facilities,on=['facid'],how='inner').query('starttime>="2012-09-21"').query('starttime<"2012-09-22"').query('name.str.startswith("Tennis")',engine='python')
q_2['name']
```

```python
#How can you output a list of all members, including the individual who recommended them (if any), 
#without using any joins? Ensure that there are no duplicates in the list, 
#and that each firstname + surname pairing is formatted as a column and ordered.


q_3=members.join(members,how='left',on='recommendedby',rsuffix='_y')
q_3["member"]=q_3['firstname']+" "+q_3['surname']
q_3['recommender']=q_3['firstname_y']+" "+q_3['surname_y']
q_3[['member','recommender']].sort_values('member')
```

```python
#Produce a count of the number of recommendations each member has made. Order by member ID.
members.groupby(by=['recommendedby']).size()
```

```python
#Produce a list of the total number of slots booked per facility. For now, 
#just produce an output table consisting of facility id and slots, sorted by facility id.
q_4=bookings.groupby(by=['facid']).sum()
q_4['slots']
```

```python
#Produce a list of the total number of slots booked per facility in the month of September 2012. 
#Produce an output table consisting of facility id and slots, sorted by the number of slots.

q_5=bookings.query('starttime>="2012-09-01"').query('starttime<"2012-10-01"').groupby(by=['facid']).sum()
q_5['slots'].sort_values()
```

```python
#Produce a list of the total number of slots booked per facility per month in the year of 2012. 
#Produce an output table consisting of facility id and slots, sorted by the id and month.
q_6=bookings
q_6['month']=pd.DatetimeIndex(q_6['starttime']).month
q_6[['facid','slots','month']].groupby(by=['facid','month']).sum().sort_values(by=['facid'])
```

```python
#Find the total number of members (including guests) who have made at least one booking.
bookings.groupby('memid')['memid'].nunique().sum()
```

```python
#Produce a list of each member name, id, and their first booking after September 1st 2012. Order by member ID.
q_7=bookings.query('starttime>="2012-09-01"').merge(members,how="left",on=['memid'])
q_7[['surname','firstname','memid','starttime']].groupby(by=['surname','firstname','memid']).min().sort_values(by=['memid'])
```

```python
#Output the names of all members, formatted as 'Surname, Firstname'
q_8=members
q_8['name']=q_8['surname']+', '+q_8['firstname']
q_8['name']
```

```python
#Perform a case-insensitive search to find all facilities whose name begins with 'tennis'. Retrieve all columns.
facilities.query('name.str.startswith("Tennis")',engine='python')
```

```python
#You've noticed that the club's member table has telephone numbers with very inconsistent formatting.
#You'd like to find all the telephone numbers that contain parentheses, 
#returning the member ID and telephone number sorted by member ID.

q_10=members.query('telephone.str.startswith("(")',engine='python')
q_10[['memid','telephone']]
```

```python
#You'd like to produce a count of how many members you have whose surname starts with each letter of the alphabet. 
#Sort by the letter, and don't worry about printing out a letter if the count is 0.
members.groupby(members['surname'].str.slice(0,1)).size().reset_index(name='count')
```
